```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 设置 matplotlib 中文字体
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号

# 1. 加载数据
fedavg_data = pd.read_csv('FedAvg MNIST test_accuracy 2.csv', header=None, names=['Accuracy']).iloc[1:,:]
moon_data = pd.read_csv('MOON MNIST test_accuracy 2.csv', header=None, names=['Accuracy']).iloc[1:,:]

# 绘制轮次
max_rounds = 100
fedavg_data = fedavg_data.iloc[:max_rounds]
moon_data = moon_data.iloc[:max_rounds]

# 2. 计算最高准确率
fedavg_max = fedavg_data['Accuracy'].max()
moon_max = moon_data['Accuracy'].max()

fedavg_max_round = fedavg_data['Accuracy'].idxmax()  
moon_max_round = moon_data['Accuracy'].idxmax()

# 3. 绘制曲线
plt.figure(figsize=(10, 6))

# FedAvg 曲线
plt.plot(fedavg_data['Accuracy'], label=f'FedAvg', 
         color='blue', linestyle='--', linewidth=2, marker='o', markersize=5)

# MOON 曲线
plt.plot(moon_data['Accuracy'], label=f'MOON', 
         color='red', linestyle='-', linewidth=2, marker='s', markersize=5)

# 标注最高点
plt.scatter(fedavg_max_round, fedavg_max, color='blue', s=100, zorder=5)  
plt.scatter(moon_max_round, moon_max, color='red', s=100, zorder=5)

# 图表装饰
plt.title('MNIST 测试准确率对比: FedAvg vs MOON', fontsize=14)
plt.xlabel('通信轮次', fontsize=12)
plt.ylabel('测试准确率', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.6)
plt.legend(fontsize=12, loc='lower right')
plt.ylim(0, max(fedavg_max,moon_max)*1.1)  
plt.xlim(0, max_rounds+1)  

# 设置x轴刻度
plt.xticks(np.arange(0, max_rounds+1, 5))

# 显示图表
plt.tight_layout()
plt.show()

# 4. 打印最高准确率
print(f"FedAvg 最高测试准确率: {fedavg_max:.4f} (轮次 {fedavg_max_round})")
print(f"MOON 最高测试准确率: {moon_max:.4f} (轮次 {moon_max_round})")
```
